```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

from sklearn import multioutput
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, r2_score

from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, GRU, Dropout, RepeatVector, TimeDistributed, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping


df = pd.read_csv('df_for_learning.csv')
df.tail(10)
```

```python
df.shape, df.columns.tolist()
```

```python
featured_cols = [
    'vol_mean_288',
    'alt_mean_288',
    'zscore_12',
    'cap_mean_12',
    'cap_mean_24',
    'cap_mean_288',
    'cap_std_288',
    'log_market_pairs',
    'log_total_cap',
    'log_altcoin_cap',
    'volume_ratio',
    'vol_mean_24',
    'dow_sin'
]

target_cols = [f'y{i}' for i in range(1, 49)]
df = df.dropna(subset=featured_cols + target_cols).copy()
```

```python
def prepare_data(df, featured_cols=featured_cols, target_cols=target_cols):
    X = df[featured_cols].to_numpy()
    y = df[target_cols].to_numpy()

    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=0.2,
        shuffle=False
    )

    scaler_x = MinMaxScaler()
    scaler_y = MinMaxScaler()

    X_train_scaled = scaler_x.fit_transform(X_train)
    y_train_scaled = scaler_y.fit_transform(y_train)

    X_test_scaled = scaler_x.transform(X_test)
    y_test_scaled = scaler_y.transform(y_test)

    return X_train_scaled, y_train_scaled, X_test_scaled, y_test_scaled, scaler_x, scaler_y

def create_sequenes(X, y, n_lags=48, horizon=48):
    Xs, ys = [], []
    for i in range(len(X) - n_lags - horizon):
        Xs.append(X[i:i + n_lags])
        ys.append(y[i + n_lags])

    return np.array(Xs), np.array(ys)
```

```python
def build_gru_seq2seq(n_lags, n_features, horizon, units=128, dropout=0.3):
    encoder_inputs = Input(shape=(n_lags, n_features))
    encoder = GRU(units=units, dropout=dropout, return_sequences=False)(encoder_inputs)
    encoder_output = Dense(units=units // 2, activation='relu')(encoder)

    decoder_input = RepeatVector(horizon)(encoder_output)
    decoder = GRU(units=units, dropout=dropout, return_sequences=True)(decoder_input)
    decoder_output = TimeDistributed(Dense(1))(decoder)

    model = Model(encoder_inputs, decoder_output)
    model.compile(optimizer='adam', loss='mae')
    return model
```

```python
X_train_scaled, y_train_scaled, X_test_scaled, y_test_scaled, scaler_x, scaler_y = prepare_data(df, featured_cols, target_cols)

X_train_seq, y_train_seq = create_sequenes(X_train_scaled, y_train_scaled)
X_test_seq, y_test_seq = create_sequenes(X_test_scaled, y_test_scaled)

y_train_seq = y_train_seq.reshape(-1, 48, 1)
y_test_seq = y_test_seq.reshape(-1, 48, 1)

model = build_gru_seq2seq(n_lags=48, n_features=X_train_seq.shape[2], horizon=48, units=128, dropout=0.3)

early_stopping = EarlyStopping(patience=10, restore_best_weights=True)
history = model.fit(
    X_train_seq, y_train_seq,
    validation_split=0.2,
    epochs=50,
    batch_size=32,
    callbacks=[early_stopping],
    verbose=1
)
```

```python
def pred_n_metrics(X_test_seq, y_test_seq, model, scaler_y):
    y_pred_scaled = model.predict(X_test_seq)
    y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 48))
    y_true = scaler_y.inverse_transform(y_test_seq.reshape(-1, 48))

    mae_per_horizon = [mean_absolute_error(y_true[:,i], y_pred[:, i]) for i in range(48)]
    r2_per_horizon = [r2_score(y_true[:,i], y_pred[:, i]) for i in range(48)]

    print(f"Mean MAE: {np.mean(mae_per_horizon):.4f}")
    print(f"Mean R²: {np.mean(r2_per_horizon):.4f}")

    return y_pred, y_true

def visualization(y_true, y_pred, model_name):
    i = -1
    plt.figure(figsize=(12, 5))
    plt.plot(y_true[i], label='Actual')
    plt.plot(y_pred[i], label='Predicted')
    plt.title(f'{model_name} Volume Prediction On 48 Steps (240 minutes)')
    plt.xlabel('Prediction Step')
    plt.ylabel('log(volume_24h)')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
y_pred, y_true = pred_n_metrics(X_test_seq, y_test_seq, model, scaler_y)
visualization(y_true, y_pred, model_name='GRU')
```
